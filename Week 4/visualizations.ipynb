- What is the distribution of the `category_name` column?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('most_funded_feb_2023.csv')
category_counts = df['category_name'].value_counts()

plt.figure(figsize=(8, 6))
x_axis = np.linspace(0,30, num=30)
y_axis = x_axis
plt.bar(category_counts.index, category_counts.values)
plt.xlabel(r'Category Name', fontname="Times New Roman", fontsize = 20)
plt.ylabel(r'Count', fontname = 'Times New Roman', fontsize = 20)
plt.title('Distribution of Category Names',  fontsize=10)
plt.xticks(rotation=90, fontsize=5)
plt.tight_layout()
plt.grid(True)
plt.show()




```

#### *According to the data visually represented above, we see that Tabletop Games ranks first with over 700 among the rest of the categories. Closely follwed by Product Design which is just under the 600 mark.*

- Which country has the most projects?

```python
df = pd.read_csv('most_funded_feb_2023.csv')
country_counts = df['country'].value_counts()

plt.figure(figsize=(8, 6))
x_axis = np.linspace(0,30, num=30)
y_axis = x_axis
plt.stem(country_counts.index, country_counts.values)
plt.xlabel(r'Country', fontname="Times New Roman", fontsize = 20)
plt.ylabel(r'Number of Projects', fontname="Times New Roman", fontsize = 20)
plt.title('Number of Projects by Country', fontname="Times New Roman", fontsize = 20)
plt.tight_layout()
plt.grid(True)
plt.show()
```

#### *According to the data visually represented above, we see that the United States far surpasses every other Country by far. This is in part due to the fact that the United States has 1750 more projects then the next remotley closest Country.*

- Which parent category has the most funded projects?

```python
df = pd.read_csv('most_funded_feb_2023.csv')
parent_category_counts = df['category_parent_id'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(8, 6))
x_axis = np.linspace(0,20, num=30)
y_axis = x_axis
plt.bar(parent_category_counts.index, parent_category_counts.values)
plt.xlabel(r'Parent Category ID', fontname="Times New Roman", fontsize = 20)
plt.ylabel(r'Count', fontname="Times New Roman", fontsize = 20)
plt.title('Most Funded Parent Categories', fontname="Times New Roman", fontsize = 20)
plt.tight_layout()
plt.grid(True)
plt.show()
```

- Create a user-defined function that allows a user to input a category and returns a visualization of your choice.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('most_funded_feb_2023.csv')
random_category = df['category'].sample().values[0]
category_df = df[df['category'] == random_category]
project_counts = category_df['project_name'].value_counts().nlargest(10)
    
plt.figure(figsize=(10, 6))
plt.stem(project_counts.index, project_counts.values)
plt.title(f"Top 10 Most Funded Projects in the {random_category} Category")
plt.xlabel("Project Name")
plt.ylabel("Number of Fundings")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
