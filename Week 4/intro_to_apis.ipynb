### Problem 1:
Check out this Pokemon API https://pokeapi.co/
Use the requests package to connect to this API and get and store data for 5 different pokemon.
Get the pokemons: name, atleast one ability's name, base_experience, and the URL for its sprite (an image that shows up on screen) for the 'front_shiny', attack base_state, hp base_stat, defense base_stat

**BONUS POINTS IF YOU CAN POPULATE THE IMAGE!**

```python
import requests
import random

def get_pokemon_data(pokemon_name):
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name}"
    response = requests.get(url)
    if response.ok:
        return response.json()
    else:
        return None

def get_pokemon_info(pokemon_data):
    name = pokemon_data['name']
    ability_name = pokemon_data['abilities'][0]['ability']['name']
    base_experience = pokemon_data['base_experience']
    sprite_url = pokemon_data['sprites']['front_shiny']
    attack_stat = pokemon_data['stats'][1]['base_stat']
    hp_stat = pokemon_data['stats'][0]['base_stat']
    defense_stat = pokemon_data['stats'][2]['base_stat']

    return name, ability_name, base_experience, sprite_url, attack_stat, hp_stat, defense_stat

def build_team():
    pokemon_list = []
    team_size = 5

    while len(pokemon_list) < team_size:
        pokemon_id = random.randint(1, 1000)
        pokemon_data = get_pokemon_data(pokemon_id)
        if pokemon_data:
            pokemon_info = get_pokemon_info(pokemon_data)
            pokemon_list.append(pokemon_info)

    return pokemon_list

team = build_team()

for pokemon_info in team:
    name, ability_name, base_experience, sprite_url, attack_stat, hp_stat, defense_stat = pokemon_info

    print(f'\nName: {name}')
    print(f'Ability: {ability_name}')
    print(f'Base Experience: {base_experience}')
    print(f'Sprite URL: {sprite_url}')
    print(f'Attack Base Stat: {attack_stat}')
    print(f'HP Base Stat: {hp_stat}')
    print(f'Defense Base Stat: {defense_stat}')

```

### Problem 2:

I want to build the best Pokemon team I can and I need your help to do so! Using the API, build me a team of 6 pokemon who's base-stats are above 60. Return any additional information you think is needed to help understand why you chose that pokemon for the team!

```python
import requests
import random

def get_pokemon_data(pokemon_name):
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name}"
    response = requests.get(url)
    if response.ok:
        return response.json()
    else:
        return None

def build_team():
    pokemon_list = []
    base_stat_min = 60
    team_size = 6

    while len(pokemon_list) < team_size:
        pokemon_name = random.randint(1, 1000)
        pokemon_data = get_pokemon_data(pokemon_name)
        if pokemon_data:
            base_stats = pokemon_data['stats']
            if all(stat['base_stat'] > base_stat_min for stat in base_stats):
                pokemon_list.append(pokemon_data)

    return pokemon_list

team = build_team()

for pokemon_data in team:
    name = pokemon_data['name']
    types = [t['type']['name'] for t in pokemon_data['types']]
    abilities = [a['ability']['name'] for a in pokemon_data['abilities']]

    print(f'\nName: {name}')
    print(f'Type(s): {", ".join(types)}')
    print(f'Abilities: {", ".join(abilities)}')
```

### Problem 3:

Create a function that does the following:
* Input: pokemon_name(user_defined) -> str
* Output: list object containing a dictionary of values from the response object.
* Please include the type, the most current raritey level, at least one held item, and all stats

```python
import requests

def get_pokemon_details(pokemon_name):
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name}"
    response = requests.get(url)
    if response.ok:
        data = response.json()
        details = []

        types = [type_info['type']['name'] for type_info in data['types']]
        details.append({'Type': types})

        base_experience = data['base_experience']
        details.append({'Base Experience': base_experience})

        held_items = [item['item']['name'] for item in data['held_items']]
        details.append({'Held Item': held_items[0]})

        stats = {stat['stat']['name']: stat['base_stat'] for stat in data['stats']}
        details.append({'Stats': stats})

        return details
    else:
        return None

def get_pokemon_info(pokemon_name):
    pokemon_details = get_pokemon_details(pokemon_name)
    if pokemon_details:
        for detail in pokemon_details:
            for key, value in detail.items():
                print(f"{key}: {value}")
            print()
    else:
        print(f'Not a Pokémon: {pokemon_name}.')

pokemon_name = input("Enter a Pokémon name: ")
get_pokemon_info(pokemon_name)


```
