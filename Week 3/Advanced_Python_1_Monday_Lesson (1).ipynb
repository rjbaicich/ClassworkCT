### Using Regular Expressions <br>
<p>Regular expressions give us the ability to search for patterns within text, strings, files, etc. They serve several uses, such as; security measures, searching, filtering, pattern recognition, and more...</p>

##### re.compile()

```python

#using compile method, pre determines the string that we want to search for in regular experession methods

pattern = re.compile('abcd')
```

##### re.match()

```python
match = pattern.match('abcd123')
print(match.span())
```

##### re.findall()

```python
help(re.findall)
finders = pattern.findall('abcd123')
print(finders)
print([('abcd'), ('abcd123'), ('abcd')])
```

##### re.search()

```python
help(re.search)

random_string = '123 123 234 abc abcd'
search = pattern.search(random_string)#search string for pattern
print(search)
span = search.span()
print(span, type(span))
#if i know i can index it we can grab
# match from resulting string either by hard code(non reproducible)
print(random_string[16:20])#16 is always 16 and 20 is 20 no matter if string changes
#this is reproducible 
print(random_string[span[0]:span[1]])
```

##### Integer Ranges

```python
#it will search for values in string object where they match
# beginning with value 0-4 the next with 7-9 and last with 0-3

pattern_int = re.compile('[0-4][7-9][0-3]')
int_str = '67383 383'
random_numbers = pattern_int.search(int_str)
print(random_numbers)
#remember a search returns a match object the match object finds first instance of a match 
#it does not search entire string and return all object
#fiond matching num,bers from match object and use dynamic programming to return them

span = random_numbers.span()
print(int_str[span[0]:span[1]])
```

```python
#how can we use this to iterate over integer values instead of a string 

pattern_int = re.compile('[0-4][7-9][0-3]')
int_value = 67383
random_numbers = pattern_int.search(str(int_value))
print(random_numbers)
span = random_numbers.span()
print(int(str(int_value)[span[0]:span[1]]))
```

##### Character Ranges

```python
#just as we l;ooked through 
#integer values using range we can do same
# for characters in a string 
char_pattern = re.compile('[A-Z][a-z]')
my_str = 'Hello There Mr.Anderson'

# .findall() object. this searches the string for ALL the 
#matching characters ands returns a list of values
found = char_pattern.findall(my_str)
print(found)
print([('He'), ('The'), ('Mr'), ('An')])
```

### Counting Occurences

##### {x} - something that occurs {num_of_times}

```python
char_pattern_count = re.compile('[A-Z][a-z][0-3]{4}')
found_count = char_pattern_count.findall('Hello Mr. An3333derson')
print(found_count)
```

##### {x, x} - something that occurs between x and x times

```python
random_pattern = re.compile('m{1,5}')
random_statement = random_pattern.findall('m mm mmmmm this was a great soup! I am in love with Tomato soup!')
print(random_statement)
```

##### ? - something that occurs 0 or 1 time

```python
pattern = re.compile('Mrss?')

found_pat = pattern.findall('Hello M the Mr. Anderson. How is Mrs. Anderson and Ms. Anderson?')
print(found_pat)
```

##### * - something that occurs at least 0 times

```python
pattern_m = re.compile('M*s')
found_m = pattern_m.findall('MMMs name is Ms.Smith. This is Msssss')
print(found_m)
```

##### + - something that occurs at least once

```python
pattern_again = re.compile('M+s')
found_patt = pattern_again.findall('My nameis Ms. Smith. This is MMMMMMMssssss')
print(found_patt)
```

### Escaping Characters

##### \w - look for any Unicode character<br/>\W - look for anything that isnt a Unicode character

[History on Unicode](http://unicode.org/standard/WhatIsUnicode.html)

[More on Unicode Characters](https://en.wikipedia.org/wiki/List_of_Unicode_characters)

```python
print('\U0001f525')
```

```python
#w or W allows us to selct items that are unicode
#W selects anything that isnt a unicode character
my_str = 'This is a sentence. With an, exclamation point at the end!'
pattern_uni = re.compile('[\w]+')
found_uni = pattern_uni.findall(my_str)
print(found_uni)

#what if i wanted everything that wasnt a unicode character

pattern_not_uni = re.compile('[\W]+')
found_not_uni = pattern_not_uni.findall(my_str)
print(found_uni)
```

##### \d - look for any digit 0-9<br/>\D - look for anything that isnt a digit

```python
date = 'Today is the 15th. In 16 days it will be the 1st.'
pattern_num = re.compile('\d{1,2}[\D]{2}')
found_date = pattern_num.findall(date)
print(found_date)
```

### In-Class Exercise #3 <br>
<p>Print each persons name and twitter handle, using groups, should look like:</p>
<p>==============<br>
   Full Name / Twitter<br>
   ==============</p>
Derek Hawkins / @derekhawkins

 Erik Sven-Osterberg / @sverik

 Ryan Butz / @ryanbutz

 Example Exampleson / @example

 Ripal Pael / @ripalp

 Darth Vader / @darthvader

```python
import re

filename = 'names (1).txt'
pattern = r'^([^@\n]+)\s+([^@\n]+)\s+([^@\n]+@(?:\w+\.)+\w+)\s+([^@\n]+)?\s+(@\w+)?'

with open(filename, 'r') as file:
    text = file.read()
    matches = re.findall(pattern, text)

for match in matches:
    full_name = f"{match[1]} {match[0]}".strip()
    twitter_handle = match[4].strip() if match[4] else ""
    print(f'{full_name} / {twitter_handle}')



```

### Regex project

Use python to read the file regex_test.txt and print the last name on each line using regular expressions and groups (return None for names with no first and last name, or names that aren't properly capitalized)
##### Hint: use with open() and readlines()

```python
"""
Expected Output
Abraham Lincoln
Andrew P Garfield
Connor Milliken
Jordan Alexander Williams
None
None
"""
```

```python
import re

names = [
    "Abraham Lincoln",
    "Andrew P Garfield",
    "Connor Milliken",
    "Jordan Alexander Williams",
    "Madonna",
    "programming is cool"
]

pattern = r'\b[A-Z][a-zA-Z-]+\b$'

for name in names:
    match = re.search(pattern, name)
    if match and name != "Madonna":
        last_name = match.group()
        print(last_name)
    else:
        print("None")

```
